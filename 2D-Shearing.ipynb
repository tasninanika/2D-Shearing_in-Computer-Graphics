```python
import matplotlib.pyplot as plt

# Step 1: Original square points
square = [(1, 1), (1, 4), (4, 4), (4, 1)]

# Step 2: Shearing factors
shear_x = 1.0  # shear along X (horizontal slant)
shear_y = 0.0  # shear along Y (vertical slant)

# Step 3: Apply shearing
sheared_square = []
for x, y in square:
    new_x = x + shear_x * y
    new_y = y + shear_y * x
    sheared_square.append((new_x, new_y))

# Step 4: Close the shapes
square.append(square[0])
sheared_square.append(sheared_square[0])

# Step 5: Separate x and y for plotting
ox, oy = zip(*square)
sx, sy = zip(*sheared_square)

# Step 6: Plot
plt.plot(ox, oy, 'go-', label='Original Square')
plt.plot(sx, sy, 'o-', label='Sheared Square')

# Extras
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("2D Shearing of a Square")
plt.legend()
plt.axis('equal')
plt.show()
```
